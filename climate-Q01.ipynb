```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os

from pathlib import Path

# Small style adjustments for more readable plots
plt.style.use("seaborn-whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["font.size"] = 14
```

```python
# Load data
DATA_DIR = Path.home()/Path('shared/climate-data')

co2_file = DATA_DIR / "monthly_in_situ_co2_mlo_cleaned.csv"
data = pd.read_csv(co2_file, na_values=-99.99).dropna()
```

```python
# Plot using sns
fig = plt.figure(figsize=(8,6))
ax  = fig.add_subplot(111)

ax = sns.lineplot(data=data,x='month',y='c02',hue="year")
ax.legend_.set_title(None)
leg = ax.legend(loc = 'upper left', bbox_to_anchor = (1.03, 1))
```

```python
# Save figure
fig.savefig('outputs/climate-Q01.png', bbox_inches='tight')
```
